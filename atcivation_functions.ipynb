```python
import torch
import torch.nn.functional as F
from torch.autograd import Variable
import matplotlib.pyplot as plt



x = torch.linspace(-5, 5, 200)
x = Variable(x)
x_np = x.data.numpy()

y_relu = torch.relu(x).data.numpy()
y_sigmoid = torch.sigmoid(x).data.numpy()
y_tanh = torch.tanh(x).data.numpy()
y_softplus = F.softplus(x).data.numpy()


# plt to visualize these activation function
plt.figure(1, figsize=(12, 8))
plt.subplot(221)
plt.plot(x_np, y_relu, c='green', label='torch.relu')
plt.ylim((-1, 5))
plt.grid()
plt.legend(loc='best')

plt.subplot(222)
plt.plot(x_np, y_sigmoid, c='green', label='torch.sigmoid')
plt.ylim((-0.2, 1.2))
plt.grid()
plt.legend(loc='best')

plt.subplot(223)
plt.plot(x_np, y_tanh, c='green', label=' torch.tanh')
plt.ylim((-1.2, 1.2))
plt.grid()
plt.legend(loc='best')

plt.subplot(224)
plt.plot(x_np, y_softplus, c='green', label='torch.nn.functional.softplus')
plt.ylim((-0.2, 6))
plt.grid()
plt.legend(loc='best')

plt.show()
```
